## Leading Death Causes

```python
df = pd.read_csv("leading_death_causes.csv")
df.head()
```

```python
df['Cause Name'].unique()
```

```python
df = df[df['Cause Name'].isin(["Alzheimer's disease",
       'Stroke', 'CLRD', 'Diabetes', 'Heart disease',
       'Influenza and pneumonia', 'Suicide', 'Cancer', 'Kidney disease'])]
```

```python
df.Deaths = df.Deaths.str.replace(',', '')
```

```python
df['Deaths'] = pd.to_numeric(df['Deaths'])
```

```python
df.dtypes
```

```python
df[df["State"]=='United States'].sort_values(by=['Deaths'], ascending=False)['Cause Name'].unique()
```

### Top 5 Leading causes:

- 'Heart disease', 
- 'Cancer', 
- 'Stroke', 
- 'CLRD'
- "Alzheimer's disease"

```python
df = df[df['Cause Name'].isin(['Heart disease', 'Cancer', 'Stroke', 'CLRD', "Alzheimer's disease"])]
```

```python
df.head()
```

```python
df.shape
```

```python
alz = df[df['Cause Name'].isin(["Alzheimer's disease"])]
heart = df[df['Cause Name'].isin(['Heart disease'])]
cancer = df[df['Cause Name'].isin(['Cancer'])]
stroke = df[df['Cause Name'].isin(['Stroke'])]
resp = df[df['Cause Name'].isin(["CLRD"])]
```

```python
alz.head()
```

```python
alz.rename({'Deaths': "Alzheimer's"}, axis=1, inplace=True)
alz.drop(["113 Cause Name","Cause Name","Age-adjusted Death Rate"], axis='columns', inplace=True)
alz.reset_index(drop=True, inplace=True)
alz.head()
```

```python
heart.rename({'Deaths': "Heart"}, axis=1, inplace=True)
heart.drop(["113 Cause Name","Cause Name","Age-adjusted Death Rate"], axis='columns', inplace=True)
heart.reset_index(drop=True, inplace=True)
heart.head()
```

```python
cancer.rename({'Deaths': "Cancer"}, axis=1, inplace=True)
cancer.drop(["113 Cause Name","Cause Name","Age-adjusted Death Rate"], axis='columns', inplace=True)
cancer.reset_index(drop=True, inplace=True)
cancer.head()
```

```python
stroke.rename({'Deaths': "Stroke"}, axis=1, inplace=True)
stroke.drop(["113 Cause Name","Cause Name","Age-adjusted Death Rate"], axis='columns', inplace=True)
stroke.reset_index(drop=True, inplace=True)
stroke.head()
```

```python
resp.rename({'Deaths': "Respiratory"}, axis=1, inplace=True)
resp.drop(["113 Cause Name","Cause Name","Age-adjusted Death Rate"], axis='columns', inplace=True)
resp.reset_index(drop=True, inplace=True)
resp.head()
```

```python
m = pd.merge(alz, heart, on=["State","Year"])
m = pd.merge(m, cancer, on=["State","Year"])
m = pd.merge(m, stroke, on=["State","Year"])
m = pd.merge(m, resp, on=["State","Year"])
m.head()
```

```python
m = m[m['Year']==2017]
```

```python
m.to_csv('deaths.csv', index=False)
```

## Hospital Counts

```python
df = pd.read_csv('state_hospital_count.csv')
df.head()
```

```python
df["State"] = df["State"].apply(lambda x : x[5:])
```

```python
df.Hospitals = df.Hospitals.str.replace(',', '')
df['Hospitals'] = pd.to_numeric(df['Hospitals'])
df = df[df['Hospitals']>3]
df = df[df['Hospitals']<2000]
```

```python
df.at[8,'State']='District of Columbia'
```

```python
extra = list(set(df['State']) - set(resp['State']))
```

```python
df = df[~df["State"].isin(extra)].reset_index(drop=True)
```

```python
df.to_csv('hospitalcounts.csv',index=False)
```
